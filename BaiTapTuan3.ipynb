W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
n = int(input())
if n & (n - 1) == 0:
    print("YES")
else:
    print("NO")
```

[Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường". 

```python
a, b, c = map(int, input().split())

if a + b <= c or a + c <= b or b + c <= a:
    print("Khong phai tam giac")
else:
    if a == b and b == c:
        print("Tam giac deu")
    elif a == b or b == c or a == c:
        print("Tam giac can")
    else:
        print("Tam giac thuong")
```

W3A14 
[Giải phương trình bậc nhất] 
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra 
nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô 
nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.

```python
a, b = map(float, input().split())

if a == 0:
    if b == 0:
        print("Vo so nghiem")
    else:
        print("Vo nghiem")
else:
    res = -b / a
    ans = round(res, 2)
    print(ans)
```

W3A16 
[Rounding] 
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số 
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả 
cách nhau một dấu cách

```python
x = float(input())

# làm tròn xuống
down = int(x)

# làm tròn lên
if x == int(x):
    up = int(x)
else:
    up = int(x) + 1

# làm tròn gần nhất
if x - int(x) >= 0.5:
    near = int(x) + 1
else:
    near = int(x)

print(up, down, near)
```

W3A17 
Cấp số nhân  
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 
cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không?  
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội 
và so sánh vs c, c nhân công bội và so sánh vs d.  

```python
a,b,c,d = map(int,input().split())

q = b // a
if b % a == 0 and c == b * q and d == c * q:
    print("Yes")
else:
    print("No")
```
