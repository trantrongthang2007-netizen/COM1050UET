W4A2
[PrimeNumber]
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 
Kiểm tra số này có phải là số nguyên tố hay không? 


```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
n = -1
while n <= 0:
    n = int(input())
print(is_prime(n))
```

W4A6 Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."


```python
def prime(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
a,b = map(int,input().split())
s = 0
for i in range(a,b+1):
    if prime(i):
        s += i
print(s)
```

W4A8
ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 


```python
def dao(n):
    res = 0
    while n > 0:
        res = res * 10 + n % 10
        n //= 10
    return res
n = int(input())
cnt = 0
while dao(n) != 1:
    n += dao(n)
    cnt += 1
print(n)
```

W4A9
[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).


```python
# so cp chu so doi 1 khac nhau
def cp(n):
    a = {}
    x = n
    cnt = 0
    while n > 0:
        r = n % 10
        a[r] = 1
        n //= 10
        cnt += 1
    if int(x**0.5)** 2 == x and len(a) == cnt:
        return True
    return False
n = int(input())
id = 1
while id*id < n:
    if cp(id*id):
        print(id*id,end = ' ')
    id += 1
```

W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 

```python
#collatz
def collatz(n):
    cnt = 1
    while n != 1:
        cnt += 1
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
    return cnt
n = int(input())
id,Max = 0,0
for i in range(1, n + 1):
    if collatz(i) > Max:
        Max = collatz(i)
        id = i
print(id, Max)
```

W4A11
[EvenDivisors]
Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.


```python
def Count(n):
    cnt = 0
    for i in range(1,n+1):
        if n % i == 0 and n % 2 == 0:
            cnt += 1
    return cnt
n = int(input())
print(Count(n))
```

W4A13 Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false" 


```python
def tonguoc(n):
    s = 0
    for i in range(1,n):
        if n % i == 0:
            s += i
    return s
a,b = map(int,input().split())
print(tonguoc(a)==b)
```

W4A14
[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất

```python
def GCD(a,b):
    if b == 0:
        return a
    return GCD(b,a%b)
a,b = map(int,input().split())
print(GCD(a,b))
```
