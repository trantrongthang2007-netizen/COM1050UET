Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
def istriangle(a, b, c):
    return a + b > c and a + c > b and b + c > a
def type_of_triangle(a, b, c):
    if a == b == c:
        return "Tam giac deu"
    elif a == b or a == c or b == c:
        return "Tam giac can"
    elif a ** 2 + b ** 2 == c ** 2 or a ** 2 + c ** 2 == b ** 2 or b ** 2 + c ** 2 == a ** 2:
        return "Tam giac vuong"
    else:
        return "Tam giac thuong"
a, b, c = map(float, input("Nhap ba canh cua tam giac: ").split())
if istriangle(a, b, c) == False:
    print("Khong phai la ba canh cua mot tam giac")
else:
    print("Loai tam giac:", type_of_triangle(a, b, c))
p = (a + b + c) / 2
print("Dien tich tam giac la:", (p * (p - a) * (p - b) * (p - c)) ** 0.5)
print("Chu vi tam giac la:", a + b + c)
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float, input("Nhap a, b, c: ").split())
if a != 0:
    delta = b ** 2 - 4 * a * c
    if delta > 0:
        x1 = (-b + delta ** 0.5) / (2 * a)
        x2 = (-b - delta ** 0.5) / (2 * a)
        print("Phuong trinh co hai nghiem phan biet: x1 =", x1, "va x2 =", x2)
    elif delta == 0:
        x = -b / (2 * a)
        print("Phuong trinh co nghiem kep: x =", x)
    else:
        print("Phuong trinh vo nghiem")
else:
    if b == 0 and c == 0:
        print("Phuong trinh co vo so nghiem")
    elif b == 0:
        print("Phuong trinh vo nghiem")
    else:
        x = -c / b
        print("Phuong trinh co nghiem duy nhat: x =", x)
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a, b , r = map(float, input("Nhap tam va ban kinh hinh tron: ").split())
x , y = map(float, input("Nhap toa do diem: ").split())
if (x - a) ** 2 + (y - b) ** 2 < r ** 2:
    print("Diem nam trong hinh tron")
elif (x - a) ** 2 + (y - b) ** 2 == r ** 2:
    print("Diem nam tren hinh tron")
else:
    print("Diem nam ngoai hinh tron")
```
