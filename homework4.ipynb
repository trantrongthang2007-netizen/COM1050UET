W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 


```python
def UCLN(a,b):
    if a == 0:
        return b
    if b == 0:
        return a
    if a > b:
        return UCLN(a % b, b)
    else:
        return UCLN(a, b % a)
a = int(input())
b = int(input())
for i in range(UCLN(a,b)):
    if UCLN(a,b) % i == 0:
        print(i)
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
def Tong(n):
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s
n = int(input())
print(Tong(n))  
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
odd = 0
even = 0
while n > 0:    
    if (n % 10) % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10
print("So chu so le:", odd)
print("So chu so chan:", even)
```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
n = 0
Max = -1000000
Min = 100000
while n != -1:
    n = int(input())
    Max = max(Max,n)
    Min = min(Min,n)
print("Max:", Max)
print("Min:", Min)
```

W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa

```python
s = input()
ans = ""
for i in s:
    if ord("A") <= ord(i) <= ord("Z"):
        ans += chr(ord(i) + 32)
    else:
        ans += chr(ord(i) - 32)
print(ans)
```

W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
s = input()
hoa, thuong , so = 0,0,0
for i in s:
    if ord("A") <= ord(i) <= ord("Z"):
        hoa += 1
    elif ord("a") <= ord(i) <= ord("z"):
        thuong += 1
    else:
        so += 1
print("So chu cai hoa:", hoa)
print("So chu cai thuong:", thuong)
print("So chu so:", so)
```
