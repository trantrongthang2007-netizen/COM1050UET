W5A16: Số Armstrong

```python
def is_amstrong(n):
    s = str(n)
    k = len(s)
    s = 0
    for i in range(k):
        s += int(s[i]) ** k
    return s == n
```

W5A18: Số đối xứng (Palindrome)

```python
def is_palindrome(n):
    n = abs(n)
    a = n
    x = 0
    while n > 0:
        x = x * 10 + n % 10
        n //= 10
    return x == a
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.


```python
def quadratic_equation(a, b, c):
    delta = b**2 - 4*a*c
    if delta < 0:
        return "Vo nghiem"
    elif delta == 0:
        return -b / (2*a)
    else:
        x1 = (-b + delta**0.5) / (2*a)
        x2 = (-b - delta**0.5) / (2*a)
        return x1,x2
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (y3 - y2) * (x2 - x1) == (y2 - y1) * (x3 - x2)
```

W5A25 Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)
 

```python
def cross(A, B, P):
    return (B[0]-A[0])*(P[1]-A[1]) - (B[1]-A[1])*(P[0]-A[0])

def point_in_quad(quad, P):
    A, B, C, D = quad
    
    c1 = cross(A, B, P)
    c2 = cross(B, C, P)
    c3 = cross(C, D, P)
    c4 = cross(D, A, P)
    
    if (c1 >= 0 and c2 >= 0 and c3 >= 0 and c4 >= 0) or (c1 <= 0 and c2 <= 0 and c3 <= 0 and c4 <= 0):
        return True
    return False

quad = [(0,0), (4,0), (4,3), (0,3)]
P = (2,1)

print(point_in_quad(quad, P))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math
def Cal(A,B,C):
    BA = (A[0] - B[0], A[1] - B[1])
    BC = (C[0] - B[0], C[1] - B[1])
    tu = BA[0] * BC[0] + BA[1] * BC[1]
    mau = ((BA[0]**2 + BA[1]**2)**0.5) * ((BC[0]**2 + BC[1]**2)**0.5)
    if mau == 0:
        return 0
    else:
        return math.degrees(math.acos(tu / mau))
A = (1,2)
B = (3,4)
C = (5,2)
print(Cal(A,B,C))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).


```python
def gt(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * gt(n-1)
def C(n, k):
    return gt(n) // (gt(k) * gt(n-k))
```

W5A34 Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def check(n):
    a = {}
    cnt = 0
    while n > 0:
        cnt += 1
        r = n % 10
        a[r] = 1
        n //= 10
    return len(a) == cnt

n = int(input())

if n < 0:
    print("0")
else:
    for i in range(1, n+1):
        if check(i):
            print(i, end=" ") 
```
