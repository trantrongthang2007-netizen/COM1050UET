W2A3

```python

soa = int(input("Nhập số nguyên a:"))

sob = int(input("Nhập số nguyên b:"))
Tong = int(soa) + int(sob)
print("Tổng là:", Tong)
Hieu = int(soa) - int(sob)
print("Hiệu là:", Hieu)
Tich = int(soa) * int(sob)
print("Tích là:", Tich)
Phan_nguyen = int(soa) // int(sob)
print("Phần nguyên là:", Phan_nguyen)
Phan_du = int(soa) % int(sob)
print("Phần dư là:",Phan_du)
Ket_qua_chia_thuc = round (int(soa) / int(sob), 2)
print("Kết quả chia thực là:", Ket_qua_chia_thuc)
```
