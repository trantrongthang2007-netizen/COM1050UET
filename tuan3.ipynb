W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
n = int(input())
if n & (n - 1) == 0:
    print("YES")
else:
    print("NO")
```

W3A5 viêt chương trình chia một số m cho n, kết quả thu được làm tròn lên 

```python
a , b , k = map(int,input().split())
x = a / b * 10**(k + 1)
if int(x) % 10 >= 5:
    x = int(x) // 10 + 1
else:    
    x = int(x) // 10
print(x / 10**(k))
```

W3A11
[Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường". 

```python
a = [1 , 2 , 3 , 4 , 1223 , 6 , 7 , 8 , 123 , 10]
max1 = a[0]
max2 = a[1]
max3 = a[2]
for i in range(3 , len(a)):
    if a[i] > max1:
        max3 = max2
        max2 = max1
        max1 = a[i]
    elif a[i] > max2:
        max3 = max2
        max2 = a[i]
    elif a[i] > max3:
        max3 = a[i]
print(max1, max2, max3)
```
