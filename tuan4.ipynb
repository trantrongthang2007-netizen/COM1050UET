Nhập vào 1 số nguyên là số tháng trong 1 năm nhuận
In ra số ngày của tháng đó

```python
n = int(input("Tháng "))
if 1 <= n <= 12:
    if n == 1 or n == 3 or n == 5 or n == 7 or n == 8 or n == 10 or n == 12:
        print(f"Tháng {n} có 31 ngày")
    elif n == 2:
        print(f"Tháng {n} có 29 ngày")
    else:
        print(f"Tháng {n} có 30 ngày")
else:
    print(f"Tháng {n} không hợp lệ")
```

```python
#Match - case
n = int(input())
t = int(input())
match n:
    case 1|3|5|7|8|10|12:
        s = f"Tháng {n} có 31 ngày"
    case 2:
        if (t%4==0 and t%100!=0) or t%400==0:
            s = f"Tháng {n} có 29 ngày"
        else:
            s = f"Tháng {n} có 28 ngày"
    case 4|6|9|11:
        s = f"Tháng {n} có 30 ngày"
    case _:
        s = f"Tháng {n} không hợp lệ"
print(s)
```
